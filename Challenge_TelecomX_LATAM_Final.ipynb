###Importando archivo JSON

```python
import requests
import pandas as pd
```

###Cargando los datos desde url y convirti√©ndolos a un DataFrame de Pandas

```python
url ='https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

df.head()
```

```python
df.info()
```

###Normalizando los datos

```python
df_normalizado = pd.json_normalize(data)
df_normalizado.info()
```

```python
df_normalizado.head()
```

###Validando qu√© valores se encuentran en cada columna para buscar errores en los datos (l√≠neas tomadas del live)

```python
for i in df_normalizado.columns:
  print(f"Valores en la columna'{i}:{df_normalizado[i].nunique()}")
  if df_normalizado[i].nunique() < 10:
    print(df_normalizado[i].unique())
    print('*' * 50)
```

###Validando si hay valores duplicados y/o nulos (l√≠neas tomadas del live)

```python
duplicados = df_normalizado.duplicated().sum()
print(f"Cantidad de l√≠neas duplicadas: {duplicados}")

valores_nulos = df_normalizado.isnull().values.sum()
print("\nValores nulos por columna:")
print(valores_nulos[valores_nulos > 0])
```

###Validando si hay valores en blanco por columna(l√≠neas tomadas del live)

```python
valores_vacios = df_normalizado.apply(lambda x: x.astype(str).str.strip() == '').sum()
print("\nValores vacios por columna:")
print(valores_vacios[valores_vacios > 0])
```

###Reemplazamos los espacios vacios de acuerdo al informaci√≥n en la columna "account.Charges.Monthly"

```python
import numpy as np

#Para reemplazar los espacios vacios se compara con el valor en la columna
#account.Charges.Monthly, si existe valor mayor a cero se asume que el cliente
#no se ha retirado, por tanto se ingresa un valor de No

#Usando for
'''
df_normalizado2 = df_normalizado
for i in range(len(df_normalizado2)):
    if df_normalizado2.loc[i, 'Churn'] == '':
        if df_normalizado2.loc[i, 'account.Charges.Monthly'] > 0:
            df_normalizado2.loc[i, 'Churn'] = 'No'
        else:
            df_normalizado2.loc[i, 'Churn'] = 'Si'
'''

#Usando numpy y operaciones vectorizadas
df_normalizado2 = df_normalizado
df_normalizado2.loc[df_normalizado2['Churn'] == '', 'Churn'] = np.where(df_normalizado2.loc[df_normalizado2['Churn'] == '', 'account.Charges.Monthly'] > 0,'No','Si')
```

```python
valores_vacios2 = df_normalizado2.apply(lambda x: x.astype(str).str.strip() == '').sum()
print("\nValores vacios por columna:")
print(valores_vacios2[valores_vacios2 > 0])
```

###Cambiamos la columna "account.Charges.Total" por float

```python
df_normalizado2['account.Charges.Total'] = pd.to_numeric(df_normalizado2['account.Charges.Total'],errors='coerce')
print(df_normalizado2['account.Charges.Total'].dtype)
```

```python
df_normalizado2.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

###Creaci√≥n de columna "Cuentas_Diarias", tomando la columna de cargos mensuales y dividendo en 30.

```python
df_normalizado2['Cuentas_Diarias'] = df_normalizado2['account.Charges.Monthly'] / 30
df_normalizado2['Cuentas_Diarias']
```

#üìÑInforme final. An√°lisis de retiro de clientes.

## 1. G√©nero
- No se observa una diferencia significativa entre hombres y mujeres en la tasa de retiro:
  - Mujeres: 939 se retiraron / 2653 permanecen ‚Üí **~26.1%**
  - Hombres: 930 se retiraron / 2745 permanecen ‚Üí **~25.3%**
- **Conclusi√≥n**: El g√©nero no parece ser un punto de influencia significagtivo en la decisi√≥n de retiro de los cleintes.

## 2. Edad - Clientes Senior (Mayores de 60)
- Los clientes mayores de 60 a√±os presentan una **mayor tasa de retiro**:
  - Senior: 476 se retiran / 706 permanecen ‚Üí **~40.3%**
  - No senior: 1393 se retiran / 4692 permanecen ‚Üí **~22.9%**
- **Conclusi√≥n**: Los adultos mayores podr√≠an estar teniendo dificultades con temas de servicio, configuraciones, actualizaciones, etc, lo cual generar√≠a molest√≠a o frustraci√≥n que podr√≠an llevar al retiro del servicio.

## 3. Meses de Contrato (Tenure)
- Se observa un **alto volumen de retiros en los primeros meses**, especialmente en el mes 0.
- A medida que el cliente permanece m√°s tiempo, es menos probable que se retire.
- **Conclusi√≥n**: La etapa inicial es cr√≠tica para retener a los nuevos clientes.

## 4. Tipo de Contrato
- Contratos **mensuales (month-to-month)** tienen la mayor tasa de retiro: 1655 retiros.
- Contratos de **uno o dos a√±os** presentan una tasa de retiro considerablemente menor.
- **Conclusi√≥n**: Los contratos a largo plazo generan mayor retenci√≥n.

## 5. Formas de Pago
- La mayor tasa de retiro se da en quienes pagan con **electronic check**: 1071 retiros.
- M√©todos autom√°ticos (tarjeta de cr√©dito o transferencia) tienen menor abandono.
- **Conclusi√≥n**: Los m√©todos manuales pueden estar ligados al olvido de los clientes o fallos en plataformas lo cual afecta en gran medida la experiencia del cliente.

## 6. Tipo de Servicio de Internet
- **Fibra √ìptica** tiene la mayor proporci√≥n de retiros (1297 clientes).
- **DSL** muestran menos retiros.
- **Conclusi√≥n**: Puede haber problemas de calidad, precio o expectativas con la fibra √≥ptica.

# ‚úÖ Recomendaciones Generales
1. **Atenci√≥n proactiva en los primeros meses del cliente**.
2. **Migrar a contratos largos mediante beneficios claros**.
3. **Fomentar pagos autom√°ticos como opci√≥n por defecto**.
4. **Soporte adaptado a clientes senior**.
5. **Evaluar calidad percibida del servicio de fibra √≥ptica**.
