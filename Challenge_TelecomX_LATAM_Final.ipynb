###Importando archivo JSON

```python
import requests
import pandas as pd
```

###Cargando los datos desde url y convirtiéndolos a un DataFrame de Pandas

```python
url ='https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

df.head()
```

```python
df.info()
```

###**Diccionario de datos**

**customerID:** número de identificación único de cada cliente

**Churn:** si el cliente dejó o no la empresa

**gender:** género (masculino y femenino)

**SeniorCitizen:** información sobre si un cliente tiene o no una edad igual o mayor a 65 años

**Partner:** si el cliente tiene o no una pareja

**Dependents:** si el cliente tiene o no dependientes

**tenure:** meses de contrato del cliente

**PhoneService:** suscripción al servicio telefónico

**MultipleLines:** suscripción a más de una línea telefónica

**InternetService:** suscripción a un proveedor de internet

**OnlineSecurity:** suscripción adicional de seguridad en línea

**OnlineBackup:** suscripción adicional de respaldo en línea

**DeviceProtection:** suscripción adicional de protección del dispositivo

**TechSupport:** suscripción adicional de soporte técnico, menor tiempo de espera

**StreamingTV:** suscripción de televisión por cable

**StreamingMovies**: suscripción de streaming de películas

**Contract:** tipo de contrato

**PaperlessBilling:** si el cliente prefiere recibir la factura en línea

**PaymentMethod:** forma de pago

**Charges.Monthly:** total de todos los servicios del cliente por mes

**Charges.Total:** total gastado por el cliente

###Normalizando los datos

```python
df_normalizado = pd.json_normalize(data)
df_normalizado.info()
```

```python
df_normalizado.head()
```

###Validando qué valores se encuentran en cada columna para buscar errores en los datos (líneas tomadas del live)

```python
for i in df_normalizado.columns:
  print(f"Valores en la columna'{i}:{df_normalizado[i].nunique()}")
  if df_normalizado[i].nunique() < 10:
    print(df_normalizado[i].unique())
    print('*' * 50)
```

###Validando si hay valores duplicados y/o nulos (líneas tomadas del live)

```python
duplicados = df_normalizado.duplicated().sum()
print(f"Cantidad de líneas duplicadas: {duplicados}")

valores_nulos = df_normalizado.isnull().values.sum()
print("\nValores nulos por columna:")
print(valores_nulos[valores_nulos > 0])
```

###Validando si hay valores en blanco por columna(líneas tomadas del live)

```python
valores_vacios = df_normalizado.apply(lambda x: x.astype(str).str.strip() == '').sum()
print("\nValores vacios por columna:")
print(valores_vacios[valores_vacios > 0])
```

###Reemplazamos los espacios vacios de acuerdo al información en la columna "account.Charges.Monthly"

```python
import numpy as np

#Para reemplazar los espacios vacios se compara con el valor en la columna
#account.Charges.Monthly, si existe valor mayor a cero se asume que el cliente
#no se ha retirado, por tanto se ingresa un valor de No

#Usando for
'''
df_normalizado2 = df_normalizado
for i in range(len(df_normalizado2)):
    if df_normalizado2.loc[i, 'Churn'] == '':
        if df_normalizado2.loc[i, 'account.Charges.Monthly'] > 0:
            df_normalizado2.loc[i, 'Churn'] = 'No'
        else:
            df_normalizado2.loc[i, 'Churn'] = 'Si'
'''

#Usando numpy y operaciones vectorizadas
df_normalizado2 = df_normalizado
df_normalizado2.loc[df_normalizado2['Churn'] == '', 'Churn'] = np.where(df_normalizado2.loc[df_normalizado2['Churn'] == '', 'account.Charges.Monthly'] > 0,'No','Si')
```

```python
valores_vacios2 = df_normalizado2.apply(lambda x: x.astype(str).str.strip() == '').sum()
print("\nValores vacios por columna:")
print(valores_vacios2[valores_vacios2 > 0])
```

###Cambiamos la columna "account.Charges.Total" por float

```python
df_normalizado2['account.Charges.Total'] = pd.to_numeric(df_normalizado2['account.Charges.Total'],errors='coerce')
print(df_normalizado2['account.Charges.Total'].dtype)
```

```python
df_normalizado2.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

###Creación de columna "Cuentas_Diarias", tomando la columna de cargos mensuales y dividendo en 30.

```python
df_normalizado2['Cuentas_Diarias'] = df_normalizado2['account.Charges.Monthly'] / 30
df_normalizado2['Cuentas_Diarias']
```

#📄Informe final. Análisis de retiro de clientes.

## 1. Género
- No se observa una diferencia significativa entre hombres y mujeres en la tasa de retiro:
  - Mujeres: 939 se retiraron / 2653 permanecen → **~26.1%**
  - Hombres: 930 se retiraron / 2745 permanecen → **~25.3%**
- **Conclusión**: El género no parece ser un punto de influencia significagtivo en la decisión de retiro de los cleintes.

## 2. Edad - Clientes Senior (Mayores de 60)
- Los clientes mayores de 60 años presentan una **mayor tasa de retiro**:
  - Senior: 476 se retiran / 706 permanecen → **~40.3%**
  - No senior: 1393 se retiran / 4692 permanecen → **~22.9%**
- **Conclusión**: Los adultos mayores podrían estar teniendo dificultades con temas de servicio, configuraciones, actualizaciones, etc, lo cual generaría molestía o frustración que podrían llevar al retiro del servicio.

## 3. Meses de Contrato (Tenure)
- Se observa un **alto volumen de retiros en los primeros meses**, especialmente en el mes 0.
- A medida que el cliente permanece más tiempo, es menos probable que se retire.
- **Conclusión**: La etapa inicial es crítica para retener a los nuevos clientes.

## 4. Tipo de Contrato
- Contratos **mensuales (month-to-month)** tienen la mayor tasa de retiro: 1655 retiros.
- Contratos de **uno o dos años** presentan una tasa de retiro considerablemente menor.
- **Conclusión**: Los contratos a largo plazo generan mayor retención.

## 5. Formas de Pago
- La mayor tasa de retiro se da en quienes pagan con **electronic check**: 1071 retiros.
- Métodos automáticos (tarjeta de crédito o transferencia) tienen menor abandono.
- **Conclusión**: Los métodos manuales pueden estar ligados al olvido de los clientes o fallos en plataformas lo cual afecta en gran medida la experiencia del cliente.

## 6. Tipo de Servicio de Internet
- **Fibra Óptica** tiene la mayor proporción de retiros (1297 clientes).
- **DSL** muestran menos retiros.
- **Conclusión**: Puede haber problemas de calidad, precio o expectativas con la fibra óptica.

# ✅ Recomendaciones Generales
1. **Atención proactiva en los primeros meses del cliente**.
2. **Migrar a contratos largos mediante beneficios claros**.
3. **Fomentar pagos automáticos como opción por defecto**.
4. **Soporte adaptado a clientes senior**.
5. **Evaluar calidad percibida del servicio de fibra óptica**.



# 🧾 Conclusión Final
- El retiro de clientes no es aleatorio; está influenciado por la edad (Senior), el tipo de contrato, la forma de pago y la duración del servicio.
- Las variables género y método de pago no son determinantes por sí solas, pero pueden influir cuando se combinan con otras.
- Reforzar la experiencia inicial y adaptar servicios a distintos segmentos puede reducir significativamente la tasa de abandono.
